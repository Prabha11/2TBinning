```python
import os
import csv

with open('simBG_output.L2_summary') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    bins = []
    filtered_count = []
    percentage = []
    for row in csv_reader:
        print(row)
        bins.append(row[0])
        filtered_count.append(int(row[4]))
        percentage.append(float(row[5]))
        
print(bins)
print(filtered_count)
print(percentage)
```

```python
import pylab as plt

# DayOfWeekOfCall = [1,2,3]
# DispatchesOnThisWeekday = [77, 32, 42]

LABELS = ["11", "21", "22", "23", "31", "41", "51", "61","71","72"]

plt.bar(bins, filtered_count, align='center')
plt.xticks(bins, bins)
plt.xlabel("Bin")
plt.ylabel("Filtered count")
plt.title("Filtered count with bins")

plt.show()
```

```python
plt.bar(bins, percentage, align='center')
plt.xticks(bins, bins)
plt.xlabel("Bin")
plt.ylabel("Filtered percentage")
plt.title("Filtered percentage with bins")

plt.show()
```

```python
import csv

with open('simBG_view1.OFDEG') as csv_file:
    csv_reader1 = csv.reader(csv_file, delimiter=',')
    row_count_full = sum(1 for row in csv_reader1)
print("Full Dataset count - ",row_count_full)

with open('simBG_output.filtered_contiges_1st_phase') as csv_file:
    csv_reader2 = csv.reader(csv_file, delimiter=',')
    row_count_filtered = sum(1 for row in csv_reader2)
print("Filtered Dataset count - ",row_count_filtered)    

Labels = ["N/A", "filtered Dataset"]
values = [row_count_full-row_count_filtered, row_count_filtered]
colors = ['red', 'green']
explode = (0.1, 0)  # explode 1st slice
plt.pie(values, explode=explode, labels=Labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
